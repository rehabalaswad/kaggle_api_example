This notebooks shows you how to quickly incorporate the [Cooperative Patent Classification titles](https://www.kaggle.com/datasets/xhlulu/cpc-codes) to the train (or test) data of the [US Patent Phrase Matching competition](https://www.kaggle.com/c/us-patent-phrase-to-phrase-matching) using pandas.

* [Link to Competition](https://www.kaggle.com/c/us-patent-phrase-to-phrase-matching)
* [Link to the Dataset](https://www.kaggle.com/datasets/xhlulu/cpc-codes)
* [Discussion thread](https://www.kaggle.com/c/us-patent-phrase-to-phrase-matching/discussion/314306)

```python
import pandas as pd

train = pd.read_csv('../input/us-patent-phrase-to-phrase-matching/train.csv')
titles = pd.read_csv('../input/cpc-codes/titles.csv', dtype=str) 
# dtype=str is very important! It will ensure leading 0's are not removed.
```

```python
train.head()
```

```python
titles.head()
```

```python
merged_train = train.merge(titles, left_on='context', right_on='code')
merged_train.head()
```

Note: You can see that the class, subclass, group and main group have a lot of NaN data. This is normal, as the classification is hierarchical, which means that a code "A01B33/00" will have a title, but "A01B33" will also have a title, and "A01B", "A01", and "A" as well.
